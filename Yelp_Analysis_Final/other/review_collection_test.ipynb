```python
import requests
import json
import os
from requests_oauthlib import OAuth1
from state_abb import states_list

# read API keys
def get_auth():
    with open('config_secret.json') as cred:
        creds = json.load(cred)
        CONSUMER_KEY = creds['consumer_key']
        CONSUMER_SECRET = creds['consumer_secret']
        ACCESS_TOKEN = creds['token']
        ACCESS_SECRET = creds['token_secret']
        auth = OAuth1(CONSUMER_KEY,CONSUMER_SECRET,ACCESS_TOKEN,ACCESS_SECRET)
        return auth
```

```python
auth = get_auth()
clean = ("Bangalore: Restaurant', & Bar" + " " + "Fairfield").replace(':', '').replace(',', '').replace("'", '').replace('&','and')
business_id = '-'.join(clean.split(' '))
url = "https://api.yelp.com/v2/business/" + business_id
r = requests.get(url,auth=auth) 
r.text
```
